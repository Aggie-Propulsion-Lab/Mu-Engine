```python
import os
import csv
import pandas
import numpy as np
from scipy import interpolate
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
list = [file for file in os.listdir(os.getcwd()+"/Propane_test")]

data_list = []
variable_list = ["temperature", "pressure", "quality", "internal energy", "enthalpy", "entropy", "density (l)", "specific volume (l)", "internal energy (l)", "enthalpy (l)", "entropy (l)", "Cv (l)", "Cp (l)", "sound speed (l)", "joule-thomson (l)", "viscosity (l)", "thermal conductivity (l)", "density (g)", "specific volume (g)", "internal energy (g)", "enthalpy (g)", "entropy (g)", "Cv (g)", "Cp (g)", "sound speed (g)", "joule-thomson (g)", "viscosity (g)", "thermal conductivity (g)"]

for i in list:
    # appends density as the first value in a list
    density = int(i.split(".")[0])
    data_list.append([density, {}])

    # opens file and saves it as a nicely formatted list in haha 
    f = open(os.getcwd()+"/Propane_test/"+i)
    lol = pandas.read_csv(f, sep='\t', index_col=False)
    haha = lol.values.tolist()
    for j in haha:
        lol = j[0].split()
        haha[haha.index(j)] =  lol
    for j in range(len(variable_list)-1):
        data_list[list.index(i)][1][variable_list[j]] = []

        for k in range(len(haha)-1):
            data_list[list.index(i)][1][variable_list[j]].append(haha[k][j])


data_list.sort()
print(data_list[0])

# for given density: data_list[0][0]
# for list of outputs at that density: data_list[0][1]["variable"]
# variable options: 
# "temperature", "pressure", "quality", "internal energy", "enthalpy", "entropy", "density (l)", "specific volume (l)", 
# "internal energy (l)", "enthalpy (l)", "entropy (l)", "Cv (l)", "Cp (l)", "sound speed (l)", "joule-thomson (l)", "viscosity (l)", 
# "thermal conductivity (l)", "density (g)", "specific volume (g)", "internal energy (g)", "enthalpy (g)", "entropy (g)", "Cv (g)", 
# "Cp (g)", "sound speed (g)", "joule-thomson (g)", "viscosity (g)", "thermal conductivity (g)"


```

```python
# data[0] is density, data[1]["variable"] is variable results
d = []
t = []
p = []
h = []

for data in data_list:
    # density list of integers
    d_float = [float(data[0])] * len(data[1]["temperature"])

    # turns pressure, enthalpy, and temperature into list of integers
    t_float = [float(item) for item in data[1]["temperature"]]
    p_float = [float(item) for item in data[1]["pressure"]]
    h_float = [float(item) for item in data[1]["enthalpy"]]

    # appends values to the array
    d.append(d_float)
    t.append(t_float) 
    p.append(p_float) 
    h.append(h_float)


def p_func(density, enthalpy):
    i = d.index(density)
    j = h.index(enthalpy)
    return p[i][j]

d, h, p = np.array(d), np.array(h), np.array(p)
d_flat, h_flat, p_flat = d.flatten(), h.flatten(), p.flatten()


print(d)
print(h)
print(p)

print(str(len(h))+", ",str(len(h[0])))
print(str(len(d))+", ",str(len(d[0])))
print(str(len(p))+", ",str(len(p[0])))

# print(f)
# print(f(d_flat[0], h_flat[0]))
# print(p_flat[0])
# print(d)

f = interpolate.SmoothBivariateSpline(d_flat, p_flat, h_flat, kx=3, ky=3, s=0.1)
```

```python
d_mesh, p_mesh = np.meshgrid([*range(0,700,1)],[*range(0, 500, 1)])
# print(np.size(d_mesh))
# print(np.size(h_mesh))
h_interp = f.__call__(d_mesh, p_mesh, dx=0, dy=0, grid=False)
# print(p_interp)
# print(p)
```
